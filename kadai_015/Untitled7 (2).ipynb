```python
# Humanクラスの定義
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # nameとageを標準出力するメソッド
    def printinfo(self):
        print(self.name)
        print(self.age)

# Humanクラスのインスタンスを変数に代入
human = Human("侍太郎", 25)

# メソッドを呼び出す
human.printinfo()
```
