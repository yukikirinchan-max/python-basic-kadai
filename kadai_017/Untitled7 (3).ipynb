```python
# Humanクラスの定義
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # 大人かどうかを判定して標準出力するメソッド
    def check_adult(self):
        if self.age >= 20:
            print("大人です")
        else:
            print("大人ではありません")


# 人の情報をディクショナリで管理
human_data_list = [
    {"name": "侍太郎", "age": 25},
    {"name": "侍花子", "age": 18},
    {"name": "侍次郎", "age": 20}
]

# Humanクラスのインスタンスを格納するリスト
human_list = []

# 繰り返し処理でインスタンスを生成し、リストに追加
for data in human_data_list:
    human = Human(data["name"], data["age"])
    human_list.append(human)

# リストの要素数分だけcheck_adultメソッドを呼び出す
for human in human_list:
    human.check_adult()
```
