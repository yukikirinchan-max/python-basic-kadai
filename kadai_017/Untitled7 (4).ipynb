```python
# Humanクラスの定義
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # 大人かどうかを判定して、名前を含めて標準出力するメソッド
    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name} is adult.")
        else:
            print(f"{self.name} is not adult.")


# Humanクラスのインスタンスを生成する関数
def create_human_list(human_data_list):
    human_list = []
    for data in human_data_list:
        human_list.append(Human(data["name"], data["age"]))
    return human_list


# リストの要素数分だけcheck_adultメソッドを呼び出す関数
def check_all_adults(human_list):
    for human in human_list:
        human.check_adult()


# 人の情報をディクショナリで管理
human_data_list = [
    {"name": "侍太郎", "age": 25},
    {"name": "侍花子", "age": 18},
    {"name": "侍次郎", "age": 20}
]

# 関数を使って処理を実行
human_list = create_human_list(human_data_list)
check_all_adults(human_list)
```
